```r
source <- "./applications/final_dataset.csv"
```

```r
data <- xfun::in_dir(dir = path, expr = read.csv(file = source, sep = ","))
```

```r
str(data)
```

```r
x <- data$lateral_error
```

```r
options(repr.plot.width=15, repr.plot.height=11)

dens_x <- density(x)
hist(x, prob = TRUE, ylim = range(dens_x$y))
lines(dens_x, lwd = 2, col = 4)
```

```r
acf(x)
```

```r
nlargest <- 2000
y <- extract_nlargest_sample(x, n = nlargest)
dens_y <- density(y)
hist(y, prob = TRUE, ylim = range(dens_y$y))
lines(density(y), lwd = 2, col = 4)
```

```r
acf(y)
```

```r
gev_mixture_model <- suppressWarnings(estimate_gev_mixture_model_parameters(x = x, 
                                                                            block_sizes = 10:40,
                                                                            minimum_nblocks = 50,
                                                                            threshold = min(y),
                                                                            nlargest = nlargest,
                                                                            confidence_level = 0.95,
                                                                            use_extremal_index = TRUE,
                                                                            use_lower_threshold = FALSE,
                                                                            maximum_iterations = 1500,
                                                                            log_mv = TRUE,
                                                                            log_pw = TRUE,
                                                                            trace = FALSE,
                                                                            method = "MLE"))
```
